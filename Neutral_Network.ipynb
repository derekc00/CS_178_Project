```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import mltools as ml

np.random.seed(0)
%matplotlib inline

#Load and shuffke the data
X = np.genfromtxt("data/X_train.txt",delimiter=None)
Y = np.genfromtxt("data/Y_train.txt",delimiter=None)
X,Y = ml.shuffleData(X,Y)

Xtr,Xva,Ytr,Yva = ml.splitData(X,Y)
Xtr, Ytr = ml.shuffleData(Xtr, Ytr)

Xte = np.genfromtxt('data/X_test.txt',delimiter=None)

#print("ex  ", Xtr[1:3])
```

```python
def Neural(m1, m2, w1, w2, bias):
    ret = m1*w1 + m2*w2 + bias
    return sigmoid(ret)

def sigmoid(n):
    return 1/(1+np.exp(-n))

def deriv(n):
    return n*(1-n)

rp1 = np.random.random((14,16)) - 1
rp2 = np.random.random((16,1)) - 1
training_times = 10
print("random", rp1)
print("2 ", rp2)

# three layers of neural network
for i in range(training_times):
    t0 = Xtr
    t1 = sigmoid(np.dot(t0, rp1))
    t2 = sigmoid(np.dot(t1, rp1))
    
    # error between output and prediction
    t2_err = Ytr - t2
    
    t2_dir = t2_err*deriv(t2)
    
    t1_err = t2_dir.dot(rp1.T)
    
    t1_dir = t1_err*deriv(t1)
    
    rp2 += t1.T.dot(t2.dir)
    rp1 += t0.T.dot(t1.dir)
```
