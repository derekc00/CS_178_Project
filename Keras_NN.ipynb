```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import mltools as ml

np.random.seed(0)
X = np.genfromtxt("data/X_train.txt",delimiter=None)
Y = np.genfromtxt("data/Y_train.txt",delimiter=None)
X,Y = ml.shuffleData(X,Y)
```

```python
model = Sequential()
model.add(Dense(21, input_dim=14, activation='relu'))
model.add(Dense(14, activation='relu', kernel_initializer="uniform"))
model.add(Dense(1, activation='sigmoid', kernel_initializer="uniform"))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, epochs=5, batch_size=10, verbose=0)
# evaluate the model
scores = model.evaluate(X, Y)
```

```python
Xte = np.genfromtxt('data/X_test.txt',delimiter=None)
# calculate predictions
prediction = model.predict(Xte)
# round predictions
output = [round(x[0]) for x in prediction]
#output prediction
print(output)
```
